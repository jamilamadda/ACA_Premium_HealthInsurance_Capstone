### Filtering the dataset to one insurer,one group-Young Family and all the 3 tiers-Gold,Silver,Bronze .

```python
UHC_AllPlans = (
    (insurance['Issuer Name'] == 'UnitedHealthcare') &
    ((insurance['Metal Level'].isin(['Gold', 'Silver']) &
            (insurance['Adult Dental'] == 'Y') &
            (insurance['Child Dental'] == 'Y'))
        |
        (insurance['Metal Level'] == 'Bronze'))
)

Couple2children_Age30 = [
    'State Code',
    'County Name',
    'Metal Level',
    'Issuer Name',
    'Couple+2 children, Age 30'
]

TennesseeCounties_UHC_AllPlans_YoungFamilyOf2 = (
    insurance.loc[UHC_AllPlans, Couple2children_Age30]
    .reset_index(drop=True))
```

```python
TennesseeCounties_UHC_AllPlans_YoungFamilyOf2
```

### Since there are multiple versions of each plan tier in every county, taking the average of the plan tier by county to produce 1 row of plan tier and county combination

```python
TennesseeCounties_UHC_AllPlans_YoungFamilyOf2_filtered = (
    TennesseeCounties_UHC_AllPlans_YoungFamilyOf2[
        ['County Name', 'Metal Level', 'Couple+2 children, Age 30']
    ].copy()
)

TennesseeCounties_UHC_AllPlans_YoungFamilyOf2_filtered[
                        'Plan Cost'] = ( 
                            TennesseeCounties_UHC_AllPlans_YoungFamilyOf2_filtered[
                            'Couple+2 children, Age 30'] )
```

```python
TennesseeCounties_UHC_AllPlans_YoungFamilyOf2_filtered.drop(
    'Couple+2 children, Age 30',axis=1,inplace=True)


TennesseeCounties_UHC_AllPlans_YoungFamilyOf2_filtered
```

```python
AvgPlanCost_By_County = (
    TennesseeCounties_UHC_AllPlans_YoungFamilyOf2_filtered
    .groupby('County Name', as_index=False)
    .agg(Avg_Plan_Cost=('Plan Cost', 'mean'))
)

AvgPlanCost_By_County['Avg_Plan_Cost'] = AvgPlanCost_By_County['Avg_Plan_Cost'].round(2)

AvgPlanCost_By_County
```

```python
statewide_avg_plan_cost = AvgPlanCost_By_County['Avg_Plan_Cost'].mean()

round(statewide_avg_plan_cost, 2)
```

```python
print(
    f"Across all Tennessee counties, the average monthly health insurance "
    f"plan cost (averaged across Bronze, Silver, and Gold) is "
    f"${round(statewide_avg_plan_cost):,}."
)
```

### Pivoting the dataframe to align Bronze, Silver, and Gold plans side by side

```python
pivot_TennesseeCounties_UHC_AllPlans_YoungFamilyOf2 = (
    TennesseeCounties_UHC_AllPlans_YoungFamilyOf2_filtered
    .pivot(index='County Name', columns='Metal Level', values='Plan Cost')
    .reset_index())
pivot_TennesseeCounties_UHC_AllPlans_YoungFamilyOf2 
```

### Calculating  absolute upgrade costs across  3 plan tiers.

```python
pivot_TennesseeCounties_UHC_AllPlans_YoungFamilyOf2['Bronze_to_Silver_$'] = pivot_TennesseeCounties_UHC_AllPlans_YoungFamilyOf2[
                                                                            'Silver'] - pivot_TennesseeCounties_UHC_AllPlans_YoungFamilyOf2[
                                                                                                                                    'Bronze']

pivot_TennesseeCounties_UHC_AllPlans_YoungFamilyOf2['Gold_to_Silver_$'] = pivot_TennesseeCounties_UHC_AllPlans_YoungFamilyOf2[
                                                                          'Silver'] - pivot_TennesseeCounties_UHC_AllPlans_YoungFamilyOf2[
                                                                            'Gold']

pivot_TennesseeCounties_UHC_AllPlans_YoungFamilyOf2['Bronze_to_Gold_$'] = pivot_TennesseeCounties_UHC_AllPlans_YoungFamilyOf2[
                                                                            'Gold'] - pivot_TennesseeCounties_UHC_AllPlans_YoungFamilyOf2[
                                                                                'Bronze']
```

### Calculating Percentage Cost Differences across 3 Plan tiers.

```python
pivot_TennesseeCounties_UHC_AllPlans_YoungFamilyOf2['Bronze_to_Silver_%'] = (
    pivot_TennesseeCounties_UHC_AllPlans_YoungFamilyOf2['Bronze_to_Silver_$'
    ] / pivot_TennesseeCounties_UHC_AllPlans_YoungFamilyOf2['Bronze'] * 100
).round(1)


pivot_TennesseeCounties_UHC_AllPlans_YoungFamilyOf2['Gold_to_Silver_%'] = (
    pivot_TennesseeCounties_UHC_AllPlans_YoungFamilyOf2['Gold_to_Silver_$'
    ] / pivot_TennesseeCounties_UHC_AllPlans_YoungFamilyOf2['Gold'] * 100
).round(1)
```

```python
pivot_TennesseeCounties_UHC_AllPlans_YoungFamilyOf2['Avg_Plan_Cost'] = (
    pivot_TennesseeCounties_UHC_AllPlans_YoungFamilyOf2[['Bronze', 'Silver', 'Gold']]
    .mean(axis=1)
    .round(2)
)
```

```python
pivot_TennesseeCounties_UHC_AllPlans_YoungFamilyOf2
```

## Findings

**Upgrading from Bronze to Silver** results in a steep premium increase of approximately 49%, representing the largest cost jump across metal tiers.

**Upgrading from Gold to Silver** results in a minimal premium increase of about 4%, indicating near price parity between these two higher tiers.

**This pricing structure suggests** that Bronze plans are positioned as low-entry options, but the cost of moving to Silver is substantial and may act as a financial barrier.

**Overall, metal tier affordability is non-linear**, with the greatest financial burden occurring at the transition from Bronze to Silver rather than between higher tiers.

## Findings

### Geography matters more than income

The scatterplot shows that counties with similar per-capita income can have very different insurance premiums.  
The lack of correlation between per-capita income and average health insurance premiums across Tennessee counties highlights that ACA premiums are geographically standardized by insurance companies.

The Insurance companies decide how much families pay based on insurance rating areas.  
These areas are priced using factors such as hospital charges, specialist networks, past claims, and how common chronic diseases are in the area.

Poorer and wealthier counties can have similar premiums if both rely on the same hospitals or specialist networks.

Rural counties may face higher premiums due to older populations, higher chronic disease prevalence, and fewer healthcare providers, even when incomes are low.

While living in a lower-premium county could reduce costs, many young families are limited by job locations, school districts, and childcare access.  
Families may pay more simply because of where they live, not because of how much they earn.

### Greater impact on young families with two children

For young families, geographic pricing has a larger financial effect: a county that is just **\$100** more expensive per month adds up to about **\$1,200** more per year for the same family and plan tier.

# Question 5) To what extent do insurance issuers differ in pricing for identical plan tiers and household profiles within the same county?

## Focus on a single county to isolate issuer-level pricing differences for the same family type.

### Filter the dataset to Williamson County

```python
Williamson_AllPlans = (
    (insurance['County Name'] == 'Williamson') )
```

```python
Young_family = [
    'Metal Level',
    'Issuer Name',
    'Couple+2 children, Age 30']

Williamson_AllPlans_YoungFamilyOf2 = (
    insurance.loc[Williamson_AllPlans, Young_family]
    .reset_index(drop=True))
```

## Calculate average premiums by issuer and metal level for same group.

```python
Williamson_AllPlans_YoungFamilyOf2['Plan_Cost'] = (
    Williamson_AllPlans_YoungFamilyOf2['Couple+2 children, Age 30'])


Williamson_AllPlanCosts_YoungFamilyOf2 = Williamson_AllPlans_YoungFamilyOf2.groupby(
    ['Issuer Name', 'Metal Level']
).agg(
    Avg_Plan_cost=('Plan_Cost', 'mean')
).reset_index()
```

## Reshape the data for plotting.

```python
plot_df = (
    Williamson_AllPlanCosts_YoungFamilyOf2
    .pivot(
        index='Issuer Name',
        columns='Metal Level',
        values='Avg_Plan_cost'))
```

## Plotting Clustered Bar Chart for visually compare pricing differences across insurers within the same county.

```python
import numpy as np
import textwrap  
import matplotlib.pyplot as plt

issuer_order = [
    'Oscar Insurance Company',
    'Ambetter of Tennessee',
    'UnitedHealthcare',
    'Cigna Healthcare',
    'BlueCross BlueShield of Tennessee'
]

plot_df = plot_df.reindex(issuer_order).fillna(0)
wrapped_labels = [textwrap.fill(label, width=15) for label in plot_df.index]

x = np.arange(len(plot_df.index))
width = 0.25

fig, ax = plt.subplots(figsize=(14, 8))

ax.bar(x - width, plot_df['Bronze'], width, label='Bronze', color='#CD7F32')
ax.bar(x,         plot_df['Silver'], width, label='Silver', color='#C0C0C0')
ax.bar(x + width, plot_df['Gold'],   width, label='Gold',   color='#FFD700')

ax.set_xticks(x)
ax.set_xticklabels(wrapped_labels, fontsize=16)
ax.set_ylabel("Average Monthly Premium ($)", fontsize=18)

ax.set_title(
    "Williamson County: Average Monthly Premiums by Issuer\n"
    "Young Family (Couple + 2 Children, Age 30)",
    fontsize=18,
    pad=20
)

plt.ylim(0, plot_df.max().max() * 1.15)
plt.tight_layout()
plt.savefig("Williamson_County_Premiums_By_Issuer.png", dpi=300, bbox_inches='tight')
plt.show()
```

## Findings

### “Prices differ” → “insurer strategy creates affordability inequality even within the same county.

For a couple with two children at age 30, average monthly premiums differ by several hundred dollars between the lowest‑ and highest‑priced issuers within each metal tier
Some carriers (for example, BlueCross BlueShield and Cigna sit at the upper end of the premium range , while others (like Oscar and Ambetter) tend to offer lower prices .

Thus certain carriers compete on lower premiums, while others appear to position themselves as higher‑priced, possibly higher‑network or brand‑reputation options.

Because all bars represent the same county and household profile, the only thing changing is the issuer and metal level, this means a family that does not comparison‑shop across issuers could overpay by hundreds of dollars per month for essentially the same coverage tier, making issuer choice nearly as important as tier choice for affordability.
                                                                                                                                                                                                                                                                                                                                        
For young families, this creates a hidden risk: without careful comparison, they may unintentionally enroll in a higher-priced issuer simply due to brand recognition or familiarity, paying significantly more each month for coverage that is actuarially similar. In practical terms, issuer choice functions as a second “pricing gate” layered on top of metal tier selection.

# Question 7) What is the relative rate of change between health insurance premiums and average wages, and how does this relationship affect affordability trends for young families over time? 

### Data from 2018 to 2026 was put together to compare how premiums change each year across metal levels for the same county, insurance company, plan, and family type.

Because each yearly dataset was very large, loading individual files into pandas and merging them caused performance issues. The files were first appended in Excel and then imported into the notebook as a single dataset.

```python
Williamson_YOY_Cigna_Allgroups=pd.read_excel(
                    r'C:\Users\91983\Documents\NSS\python\projects\ACAPremiums_HealthInsurance_Capstone\data\TN_YOY_CIGNA.xlsx')
Williamson_YOY_Cigna_Allgroups.head(2)
```

### The dataset was filtered to include only records for young family of 2 for insurer Cigna Healthcare.

```python
Williamson_YOY_Cigna_YoungFamilyAge30= Williamson_YOY_Cigna_Allgroups[['Metal Level','Year',
                                                                        'Couple+2 children, Age 30' 
                                                                        ]].copy()
```

```python
Williamson_YOY_Cigna_YoungFamilyAge30=Williamson_YOY_Cigna_YoungFamilyAge30
Williamson_YOY_Cigna_YoungFamilyAge30.head(2)
```

```python
Williamson_YOY_Cigna_YoungFamilyAge30 = Williamson_YOY_Cigna_YoungFamilyAge30.sort_values(['Metal Level', 'Year'])
```

### Calculating Year on Year Percentage increase in premiums.

```python
Williamson_YOY_Cigna_YoungFamilyAge30['YoY_Percent_Increase'] = (
    Williamson_YOY_Cigna_YoungFamilyAge30
    .groupby('Metal Level')['Couple+2 children, Age 30']
    .pct_change() * 100)

Williamson_YOY_Cigna_YoungFamilyAge30['YoY_Percent_Increase'] = Williamson_YOY_Cigna_YoungFamilyAge30['YoY_Percent_Increase'].round(0)
Williamson_YOY_Cigna_YoungFamilyAge30=Williamson_YOY_Cigna_YoungFamilyAge30.reset_index(drop=True)
Williamson_YOY_Cigna_YoungFamilyAge30
```

### Plotting Line Chart for visually compare pricing differences across metal tiers for same group and same insurer.

```python
import matplotlib.pyplot as plt

metal_colors = {
    'Bronze': '#CD7F32',
    'Silver': '#C0C0C0',
    'Gold':   '#FFD700'
}

for metal in ['Bronze', 'Silver', 'Gold']:
    
    subset = Williamson_YOY_Cigna_YoungFamilyAge30[
        Williamson_YOY_Cigna_YoungFamilyAge30['Metal Level'] == metal
    ]
    
    plt.figure(figsize=(10, 6))
    
    plt.plot(
        subset['Year'],
        subset['YoY_Percent_Increase'],
        marker='o',
        linewidth=2,
        color=metal_colors[metal]
    )
    
    for x, y in zip(subset['Year'], subset['YoY_Percent_Increase']):
        if not pd.isna(y):
            plt.text(
                x, y,
                f"{int(y)}%",
                ha='center',
                va='bottom',
                fontsize=16
            )
    
    
    plt.ylabel("YoY Premium Change (%)")
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    
    plt.savefig(
        f"UHC_Williamson_{metal}_YoY_Premium_Change.png",
        dpi=300
    )
    
    plt.show()
```

## Findings

### From 2018 to about 2022, the price for the Cigna Bronze/Silver plan stayed almost the same each year. It went up and down a little, but not a lot, because the government was helping with enhanced ACA subsidies(from the American Rescue Plan and Inflation Reduction Act) and the insurance companies were competing and adjusting, but nothing big changed in the rules. The large YOY premium increase in 2026 — particularly the sharp ~37% jump in bronze plan premiums and the sharp ~50% jump in silver plan premiums reflects broader national trends in ACA Marketplace pricing. Insurers across the country filed significantly higher premiums for 2026, driven by rising medical and drug costs, labor inflation, and uncertainty around expiration of enhanced premium tax credits at the end of 2025 so companies expect fewer healthy people to stay in the plan and raise prices to cover the people who use more care.

```python
average_wages=pd.read_excel(
                    r'C:\Users\91983\Documents\NSS\python\projects\ACAPremiums_HealthInsurance_capstone\data\average_wage.xlsx')
average_wages
```

```python
average_wages = average_wages.sort_values('Year')
```

```python
average_wages['% difference'] = average_wages['Annual Average Occupational Wage'].pct_change() * 100
average_wages = average_wages.sort_values('Year', ascending=False).round(0)
average_wages
```

```python
fig, ax1 = plt.subplots(figsize=(12, 7))

bronze_df = Williamson_YOY_Cigna_YoungFamilyAge30[Williamson_YOY_Cigna_YoungFamilyAge30['Metal Level'] == 'Bronze'].sort_values('Year')
silver_df = Williamson_YOY_Cigna_YoungFamilyAge30[Williamson_YOY_Cigna_YoungFamilyAge30['Metal Level'] == 'Silver'].sort_values('Year')
wages_sorted = average_wages.sort_values('Year')

ax1.plot(bronze_df['Year'], bronze_df['YoY_Percent_Increase'].fillna(0), color='#cd7f32', label='Bronze Plan')
ax1.plot(silver_df['Year'], silver_df['YoY_Percent_Increase'].fillna(0), color='#c0c0c0', label='Silver Plan')

ax2 = ax1.twinx()
ax2.plot(wages_sorted['Year'], wages_sorted['% difference'].fillna(0), color='red', label='Average Wages')


for df, ax in [(bronze_df, ax1), (silver_df, ax1), (wages_sorted, ax2)]:
    val_2026 = df[df['Year'] == 2026].iloc[0]
    y_col = 'YoY_Percent_Increase' if 'YoY_Percent_Increase' in df.columns else '% difference'
    y_val = val_2026[y_col]

    ax.scatter(2026, y_val, color='black', zorder=5)
    
    ax.annotate(f'{int(y_val)}%', (2026, y_val), 
                textcoords="offset points", 
                xytext=(5, 5), 
                ha='left', 
                color='black', 
                fontsize=14,
                fontweight='bold')

all_values = (
    bronze_df['YoY_Percent_Increase'].fillna(0).tolist() + 
    silver_df['YoY_Percent_Increase'].fillna(0).tolist() + 
    wages_sorted['% difference'].fillna(0).tolist()
)
y_min, y_max = min(all_values) - 5, max(all_values) + 10 # Extra padding for labels

ax1.set_ylim(y_min, y_max)
ax2.set_ylim(y_min, y_max)

ax1.set_xticks(range(2018, 2027))
ax1.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{int(x)}%'))
ax2.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{int(x)}%'))
ax1.grid(True, which='both', linestyle='--', alpha=0.5)

lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')


ax1.set_xlabel("")
ax1.set_ylabel("")
ax2.set_ylabel("")
plt.title("")

plt.tight_layout()
plt.savefig('dual_axis_chart_2026_focus.png')
plt.show()
```

### In 2026, health insurance premiums rise ~35× faster than wages—turning coverage from a safety net into a financial strain. premiums are rising multiple times faster than wages.“In 2026, health insurance premiums are projected to rise 35 times faster than wages.”When premiums rise much faster than wages, households are forced to reallocate money from other essentials placing sustained pressure on families who don’t receive employer coverage or generous subsidies thus increasing the risk of underinsurance
